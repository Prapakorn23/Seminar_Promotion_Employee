```python
import numpy as np
import matplotlib.pyplot as plt
#SVM

# ข้อมูล
categories = ['SVM \nwithout SMOTE and Feature Selection', 'SVM with SMOTE', 'SVM with Feature Selection', 'SVM \nwith SMOTE and Feature Selection']
splits = ['70:30', '80:20']

Accuracy = [[93.37, 93.54], [65.98, 54.66], [93.12, 93.42], [68.18,59.59]]
Precision = [[96.06, 95.08], [78.73, 67.14], [97.58, 98.30], [74.30,69.95]]
Recall = [[27.75,28.95], [43.78,17.99], [24.49,26.64], [55.58,33.62]]
F1_score = [[43.06,44.39], [36.27,28.41], [39.15,41.92], [63.59,45.41]]


# ตั้งค่าขนาดรูป (1000px / 300dpi = 3.33 นิ้ว)
figsize_inch = (3.5,3.5 )
dpi_setting = 300

# ตำแหน่งแกน x
x = np.arange(len(splits))
width = 0.2  # ความกว้างแท่ง

# สร้างกราฟ
fig, axs = plt.subplots(1, 4, figsize=(figsize_inch[0]*4, figsize_inch[1]), dpi=dpi_setting, sharey=True)

for i in range(4):
    axs[i].bar(x - 1.5*width, Accuracy[i], width, label='Accuracy (%)', color='dodgerblue')
    axs[i].bar(x - 0.5*width, Precision[i], width, label='Precision (%)', color='tomato')
    axs[i].bar(x + 0.5*width, Recall[i], width, label='Recall (%)', color='gold')
    axs[i].bar(x + 1.5*width, F1_score[i], width, label='F1-Score (%)', color='forestgreen')

    # ใส่ค่าเปอร์เซ็นต์บนแท่ง
    for j in range(2):
        axs[i].text(x[j] - 1.5*width, Accuracy[i][j] + 2, f"{Accuracy[i][j]:.2f}", ha='center', va='bottom', fontsize=7.8)  # ปรับขนาดฟอนต์ที่นี่
        axs[i].text(x[j] - 0.5*width, Precision[i][j] + 2, f"{Precision[i][j]:.2f}", ha='center', va='bottom', fontsize=7.8)  # ปรับขนาดฟอนต์ที่นี่
        axs[i].text(x[j] + 0.5*width, Recall[i][j] + 2, f"{Recall[i][j]:.2f}", ha='center', va='bottom', fontsize=7.8)  # ปรับขนาดฟอนต์ที่นี่
        axs[i].text(x[j] + 1.5*width, F1_score[i][j] + 2, f"{F1_score[i][j]:.2f}", ha='center', va='bottom', fontsize=7.8)  # ปรับขนาดฟอนต์ที่นี่

    axs[i].set_xticks(x)
    axs[i].set_xticklabels(splits, fontsize=10, fontweight='bold')
    axs[i].set_title(categories[i], fontsize=11, fontweight='bold')
    axs[i].set_ylim(0, 110)

# ตกแต่งเพิ่มเติม
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=4, bbox_to_anchor=(0.5, 1.12), fontsize=8)

fig.tight_layout()

# บันทึกเป็นไฟล์ (หากต้องการ)
# plt.savefig('graph.png', dpi=dpi_setting, bbox_inches='tight')

plt.show()
```
